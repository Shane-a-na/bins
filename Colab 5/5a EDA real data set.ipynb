# 'Adult' dataset

Uses 1996 USA census data "Adult" also known as "Census Income" dataset.
The Adult dataset, available at the UCI Machine Learning Repository (https://archive.ics.uci.edu/dataset/2/adult), is a widely used dataset for machine learning and data analysis tasks. It contains demographic and socio-economic information about individuals, along with a binary target variable indicating whether their income exceeds $50,000 per year.

Purpose:
The Adult dataset is often used for tasks such as classification, regression, and data exploration. It provides a realistic and diverse set of features to work with, making it suitable for various data analysis and machine learning experiments.

Source:
The dataset was extracted from the 1994 United States Census Bureau database. It is a representative sample of working-class individuals in the United States.

Data includes:

Age: The age of the individual (numeric).\
Workclass: The type of employment, such as Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, etc. (categorical).\
Education: The highest level of education achieved (categorical).\
Education-num: The numerical representation of education (numeric).\
Marital-status: Marital status, including categories such as Married-civ-spouse, Divorced, Never-married, etc. (categorical).\
Occupation: The occupation of the individual, such as Tech-support, Craft-repair, Sales, Exec-managerial, etc. (categorical).\
Relationship: The relationship status, such as Wife, Own-child, Husband, Not-in-family, etc. (categorical).\
Race: The race of the individual, including categories like White, Black, Asian-Pac-Islander, etc. (categorical).\
Sex: The gender of the individual, Male or Female (categorical).
Capital-gain: Capital gains earned (numeric).\
Capital-loss: Capital losses incurred (numeric).
Hours-per-week: The number of hours worked per week (numeric).\
Native-country: The country of origin, such as United-States, Mexico, Germany, etc. (categorical).\
Target Variable:
The target variable in the Adult dataset is "income," which represents whether an individual's income exceeds $50,000 per year. It is a binary variable, taking the values ">50K" or "<=50K."

Data Size:
The dataset contains approximately 32,000 records, with missing values represented by "?".

# Data Visualisation

## Which chart or plot graph?

Heuristics for choosing:

**Line Plot:**

Use when visualizing trends or changes over time.
Suitable for showing continuous data.
Helpful for displaying patterns and relationships between variables.
Example: Tracking stock prices over a period of time.\
**Scatter Plot:**

Use when examining the relationship between two continuous variables.
Helpful for identifying correlations and clusters.
Suitable for highlighting individual data points.
Example: Analyzing the relationship between a person's height and weight.\
**Bar Chart:**

Use when comparing different categories or groups.
Suitable for displaying discrete or categorical data.
Useful for showing comparisons, rankings, or distributions.
Example: Comparing sales performance of different products.\
**Histogram:**

Use when visualizing the distribution of a single numerical variable.
Suitable for showing the frequency or count of data within different ranges or bins.
Helpful for understanding the shape and spread of the data.
Example: Examining the distribution of exam scores among students.\
**Pie Chart:**

Use when representing parts of a whole or proportions.
Suitable for displaying categorical data and their relative sizes.
Useful for emphasizing the composition of a dataset.
Example: Showing the percentage breakdown of different types of expenses in a budget.\
These heuristics provide general guidance, but the choice of chart ultimately depends on the nature of your data and the message you want to convey. Consider the type of data, the relationships you want to explore, and the insights you want to communicate to select the most appropriate chart for your visualization.


## line plot - hrs per week by age

```python
import pandas as pd
import matplotlib.pyplot as plt
# import data
df = pd.read_csv('adult.data')

df.columns = [
'age', 'type_employer', 'fnlwgt',
'education', 'education_num', 'marital',
'occupation',' relationship', 'race',
'sex', 'capital_gain', 'capital_loss',
'hr_per_week', 'country', 'income'
]

# Group the data by 'education-num' and calculate the mean age for each education level
grouped_data = df.groupby('age')['hr_per_week'].mean()

# Convert the result to a DataFrame
line_data = grouped_data.reset_index()

# Plot the line plot
plt.plot(line_data['age'], line_data['hr_per_week'], marker='o', linestyle='-', color='b')
plt.xlabel('Age ')
plt.ylabel('Hrs per week')
plt.title('Mean hrs per week by age')
plt.grid(True)
plt.show()
```

 the average (mean) hrs per week for age  level. The result is a Pandas Series where the 'hrs per week' values are used as the index, and the corresponding values represent the mean age for each hrs per week level.

## bar chart - education
