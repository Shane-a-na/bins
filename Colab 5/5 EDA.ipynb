## line plot

Displays data points connected by straight lines, commonly used to visualize trends and changes over time. It is created using line charts or the plot() function in libraries like Matplotlib and Seaborn.  
In this example, we define two lists x and y that represent the data points for the line plot. We then use the plt.plot() function from Matplotlib to create the line plot. The first argument of plt.plot() represents the x-axis values, and the second argument represents the corresponding y-axis values.

After creating the line plot, we set labels for the x-axis and y-axis using plt.xlabel() and plt.ylabel(), respectively. We also set a title for the plot using plt.title(). Finally, we display the plot using plt.show().


```python
import matplotlib.pyplot as plt

# Data for the line plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Set plot labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

# Show the plot
plt.show()
```

##bar chart

Compares categories or groups by showing the height or length of rectangular bars. It is often used to display **categorical** data or compare discrete values. Bar charts can be created using the bar() or barplot() functions in Matplotlib and Seaborn.
In this example, we define two lists: categories representing the categories on the x-axis and values representing the corresponding values for each category. We then use the plt.bar() function from Matplotlib to create the bar chart. The first argument of plt.bar() represents the categories, and the second argument represents the values.

After creating the bar chart, we set labels for the x-axis and y-axis using plt.xlabel() and plt.ylabel(), respectively. We also set a title for the plot using plt.title(). Finally, we display the plot using plt.show().

Running this code will generate a bar chart with the given categories and values. Each category is represented by a bar, and the height of the bar represents the corresponding value.



```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 35, 15, 27, 10]

# Create a bar chart
plt.bar(categories, values)

# Set plot labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')

# Show the plot
plt.show()
```

## histogram

Displays the **distribution** of a single numerical variable by dividing it into bins and counting the frequency of values within each bin. Histograms are created using the hist() function in Matplotlib, Seaborn, or Pandas.

## random numbers

In Python, random number generation can be done using the random module, which provides various functions for generating random values. These functions utilize different algorithms to produce random numbers based on specific distributions or requirements.
In this example, we import the random module to access its functions.\
Here's a breakdown of the code:

random.**randint**(a, b) generates a random integer between a and b, inclusive.\
random.**random**() generates a random floating-point number between 0 and 1.\
random.**uniform**(a, b) generates a random floating-point number between a and b, including both ends.\
random.**choice**(sequence) selects a random element from a given sequence, such as a list or tuple.\
random.**shuffle**(sequence) randomly shuffles the elements within a sequence, modifying it in-place.\
np.random.**normal**() is used to generate random data with a normal distribution. The function takes three arguments: the mean (0 in this case), the standard deviation (1 in this case), and the number of data points (1000 in this case).\
np.random.**seed**(seed_value) used to create random data using a seed value by setting the seed of the random number generator. The seed ensures that the generated random numbers are reproducible, meaning that running the code with the same seed will produce the same sequence of random numbers.


```python
import random
import numpy as np

# Generate a random integer between 1 and 10
random_integer = random.randint(1, 10)
print("Random Integer:", random_integer)

# Generate a random floating-point number between 0 and 1
random_float = random.random()
print("Random Float:", random_float)

# Generate a random floating-point number between 0 and 10
random_float_range = random.uniform(0, 10)
print("Random Float Range:", random_float_range)

# Generate a random element from a list
my_list = ["apple", "banana", "cherry", "date", "elderberry"]
random_element = random.choice(my_list)
print("Random Element:", random_element)

# Shuffle the elements of a list randomly
random.shuffle(my_list)
print("Shuffled List:", my_list)

# Generate random data with a normal distribution
data = np.random.normal(0, 1, 10)
print("Random Data:", data)

# Set the seed for the random number generator
seed_value = 42
np.random.seed(seed_value)

# Generate random data
np_random_data = np.random.rand(5)

print("Random Data (using a seed):", np_random_data)
```

## histogram with bars and a line

We then create the histogram using plt.hist(). The bins argument specifies the number of bins for the histogram, alpha controls the transparency of the bars, and density=True normalizes the histogram to represent the probability density. We also set edgecolor='black' to add black edges to the bars for better visibility.

Next, we add a line plot on top of the histogram using plt.plot(). We create a line that represents the probability density function (PDF) of the normal distribution using the formula 1 / (np.sqrt(2 * np.pi)) * np.exp(-0.5 * (np.linspace(min(data), max(data), 100))**2). Here, we generate a range of x-values using np.linspace() and calculate the corresponding y-values based on the PDF formula.

Finally, we set the labels for the x-axis and y-axis, and provide a title for the plot. We display the plot using plt.show().

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for the histogram
data = np.random.normal(0, 1, 1000)

# Create the histogram
plt.hist(data, bins=30, alpha=0.7, density=True, edgecolor='black')

# Add a line plot on top of the histogram
plt.plot(np.linspace(min(data), max(data), 100),
         1 / (np.sqrt(2 * np.pi)) * np.exp(-0.5 * (np.linspace(min(data), max(data), 100))**2),
         color='red', linewidth=2)

# Set plot labels and title
plt.xlabel('Value')
plt.ylabel('Frequency / Probability Density')
plt.title('Histogram with Line')

# Show the plot
plt.show()
```

## density plot

a "density plot" or a "kernel density plot." It is a way to visualize the distribution of data in a smooth continuous manner, without the discrete bins used in a traditional histogram.

Instead of using bars to represent the frequencies or counts of data within specific intervals, a density plot uses a continuous line to depict the estimated probability density function (PDF) of the data distribution. The line represents the smoothed version of the distribution, providing a visual representation of the underlying shape.

In this example, we use the kdeplot() function from the Seaborn library to create the density plot. We pass the dataset data as the input, and the function estimates and plots the kernel density estimate. We then add labels and a title to the plot and display it using plt.show().

Density plots provide a smooth representation of the data distribution, making them useful for visualizing continuous data and detecting patterns or deviations from a normal distribution.


```python
import seaborn as sns

# Create an example dataset
data = [2, 3, 3, 4, 5, 6, 8, 10, 11, 12, 14]

# Create a density plot
sns.kdeplot(data)

# Set plot labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plot')

# Show the plot
plt.show()
```

## Normal distribution

A normal distribution, also known as a Gaussian distribution or a bell curve, is a statistical concept that describes a particular pattern of data distribution. It is commonly used in many fields, including statistics, mathematics, and natural sciences.

Imagine you have a large group of people, and you measure their heights. If you plot those heights on a graph, with the number of people on the vertical axis and the height on the horizontal axis, you might observe a pattern where most people have average heights, and fewer people have heights that deviate from the average. This pattern would resemble a bell-shaped curve.

A normal distribution is characterized by several key features:

Symmetry: The curve is symmetric, meaning that the data is equally distributed around the mean (average) value. The left and right sides of the curve are mirror images of each other.

Bell-shaped curve: The curve forms a smooth, bell-like shape, with the highest point located at the mean value. The curve gradually decreases as you move away from the mean in either direction.

Central tendency: The mean, median, and mode of the distribution are all equal and located at the center of the curve. This central value represents the average or typical value of the data.

Standard deviation: The spread of the data is measured by the standard deviation. It describes how much the data points vary from the mean. In a normal distribution, about 68% of the data falls within one standard deviation from the mean, 95% falls within two standard deviations, and nearly all (around 99.7%) falls within three standard deviations.

The normal distribution is widely used in statistics because it appears naturally in many real-world phenomena. It allows researchers to make predictions and estimate probabilities based on the properties of the distribution. It also serves as a basis for various statistical tests and methods.

Overall, the concept of a normal distribution provides a mathematical framework to understand and analyze patterns of data that tend to cluster around an average value with predictable levels of variation.


### example of normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random data from a normal distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution
sample_size = 1000  # Number of data points to generate
data = np.random.normal(mean, std_dev, sample_size)

# Create a histogram to visualize the data distribution
plt.hist(data, bins=30, density=True, alpha=0.7)

# Plot the probability density function (PDF) of the normal distribution
x = np.linspace(min(data), max(data), 100)
pdf = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std_dev**2))
plt.plot(x, pdf, color='red', linewidth=2)

# Set plot labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

# Show the plot
plt.show()
```

## Testing for a normal distribution

Parametric tests assume that each group is roughly normally distributed.
If the sample sizes of each group are small (n < 30), then we can use a Shapiro-Wilk test to determine if each sample size is normally distributed.
If the p-value of the test is less than a certain significance level, then the data is likely not normally distributed.

However, if the sample sizes are large then it’s better to use a Q-Q plot to visually check if the data is normally distributed.

If the data points roughly fall along a straight diagonal line in a Q-Q plot, then the dataset likely follows a normal distribution.

### Shapiro-Wilk

```python
import numpy as np
from scipy import stats
def check_normality(data):
    test_stat_normality, p_value_normality=stats.shapiro(data)
    print("p value:%.4f" % p_value_normality)
    if p_value_normality <0.05:
        print("Reject null hypothesis >> The data is not normally distributed")
    else:
        print("Fail to reject null hypothesis >> The data is normally distributed")

data = np.array([5, 14, 15, 15, 14, 19, 17, 16, 20, 22, 8, 21, 28, 11, 9, 29, 40])
check_normality(data)
```

## skewness

Skewness is a statistical measure that describes the asymmetry of a dataset's distribution. It quantifies the extent to which the data is skewed to the left or right relative to a normal distribution.

A positive skewness indicates that the dataset has a long tail on the right side, while a negative skewness indicates a long tail on the left side. A skewness value close to zero suggests a relatively symmetric distribution.
In this example, we have an array data representing a dataset. We import the skew function from scipy.stats to calculate the skewness. The skew() function takes the dataset as an input and returns the skewness value.

In this example, we start by defining an example dataset data. We then calculate the skewness using the skew() function from the scipy.stats module. Next, we create a density plot using sns.kdeplot() from the Seaborn library to visualize the data distribution. The density plot estimates and displays the kernel density estimate (KDE) of the data. Additionally, we add labels and a title to the plot, with the skewness value formatted to two decimal places. Finally, we display the plot using plt.show().

The density plot provides a smooth representation of the data distribution, and the skewness value gives an indication of the skewness of the dataset. By observing the shape of the density plot and considering the skewness value, you can gain insights into the asymmetry of the data distribution.



## right sided skew

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Create an example dataset
data = np.array([2, 3, 3, 4, 5, 6, 8, 10, 11, 12, 14])

# Calculate skewness
skewness = skew(data)

# Create a density plot
sns.kdeplot(data)

# Set plot labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plot - Skewness: {:.2f}'.format(skewness))

# Show the plot
plt.show()


```

## left side skew
In this updated example, we generate an example dataset using np.random.gamma() to create a dataset with left-skewed values. The gamma distribution is known for producing left-skewed data. We pass parameters (1, 2) to specify the shape and scale of the gamma distribution.

The remaining part of the code is similar to the previous example. We calculate the skewness of the dataset using scipy.stats.skew() and create a density plot using sns.kdeplot(). Finally, we set the plot labels and title, including the skewness value, and display the plot using plt.show().

By generating a dataset with left-skewed values and visualizing it using a density plot, this example demonstrates how to analyze and visualize a left-skewed distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Generate an example dataset with left-skewed values
data = np.random.gamma(1, 2, 1000)

# Calculate skewness
skewness = skew(data)

# Create a density plot
sns.kdeplot(data)

# Set plot labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plot - Skewness: {:.2f}'.format(skewness))

# Show the plot
plt.show()
```
