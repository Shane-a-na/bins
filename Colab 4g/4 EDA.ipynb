## Pandas describe()

The describe() function in the pandas library is used to generate descriptive statistics of a DataFrame or a Series. It provides a summary of the central tendency, dispersion, and shape of the data.

When applied to a DataFrame, the describe() function calculates statistics for each numerical column in the DataFrame. Here's a brief description of the statistics generated:

count: the number of non-null values in each column.
mean: the mean (average) value of each column.
std: the standard deviation, which measures the spread of values around the mean.
min: the minimum value in each column.
25%, 50%, 75%: the quartiles of the distribution (25th, 50th, and 75th percentiles, respectively).
max: the maximum value in each column.
These statistics provide a high-level overview of the data distribution, central tendency, and spread.



```python
import pandas as pd
df = pd.read_csv('numbers.csv')
print(df)
print(df.describe())

# Breaking down describe
description = df.describe()
mean = description.loc['mean']
print("The mean of the first column ", mean[0]) #two columns of data
print()
print("Describe for single columnn - one ")
df.describe()["One"]
```

## Basics; count, sum, min, max, range

Basic statistical measures:

Count: The count refers to the number of observations or data points in a dataset. It represents the total number of values present, excluding any missing or null values.

Sum: The sum is the result of adding up all the values in a dataset. It provides the total of all the values combined.

Minimum (Min): The minimum represents the smallest value in a dataset. It is the lowest observation present in the data.

Maximum (Max): The maximum represents the largest value in a dataset. It is the highest observation present in the data.

Range: The range is the difference between the maximum and minimum values in a dataset. It provides a measure of the spread or variability of the data.

These statistical measures help summarize and understand the characteristics of a dataset. They provide insights into the size, distribution, and overall range of the data values. They help to gain a quick understanding of the dataset's general characteristics and inform further analysis and decision-making.


```python
# count with len
nums = [99,86,87,88,111,86,103,87,94,78,77,85,86]
a = len(nums)
print(a)

# count(value, start, end)
b = nums.count(86)
print(b)

# sum
c = sum(nums)
print(c)

# min
x = min(nums)
print(x)

# max
y = max(nums)
print(y)

# range
z = y-x
print (z)
```

## Averages - measures of central tendency

Common measures of central tendency:

Mean: The mean, also known as the average, is the sum of all values in a dataset divided by the number of observations. It represents the typical value or central value of a dataset. The mean is sensitive to extreme values and can be affected by outliers.

Mode: The mode is the value that appears most frequently in a dataset. It represents the most common observation or the peak of the distribution. A dataset can have multiple modes (multimodal) if multiple values have the same highest frequency, or it can have no mode if all values occur with equal frequency.

Median: The median is the middle value of a dataset when it is arranged in ascending or descending order. It divides the dataset into two equal halves. If there is an even number of observations, the median is the average of the two middle values. The median is a robust measure of central tendency and is less affected by extreme values or outliers.

These measures provide insights into the central or typical value of a dataset and help understand its distribution.


```python
# mean (average)
import numpy

nums = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.mean(nums)

print(x)
```

```python
# median (mid-point)
import numpy

nums = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.median(nums)

print(x)
```

```python
# mode (most common)
from statistics import mode

nums = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = mode(nums)

print(x)
```

## Standard Deviations & Variance

Standard deviation and variance, two measures of dispersion or variability in a dataset:

Standard Deviation: The standard deviation is a measure of the spread or dispersion of a dataset. It quantifies how much the values deviate from the mean or average. A higher standard deviation indicates greater variability, while a lower standard deviation indicates less variability. The standard deviation is calculated by taking the square root of the variance.

Variance: Variance is another measure of the spread or dispersion of a dataset. It quantifies the average of the squared differences between each value and the mean. Variance provides a measure of how the data points are spread out from the mean value. A larger variance indicates greater variability, while a smaller variance indicates less variability.

Standard deviation and variance are important measures to understand the spread of data points within a dataset. They help assess the variability or consistency of the data, which can provide insights into the stability, reliability, or predictability of a phenomenon.


```python
# low example
import numpy  # you can use import numpy as np and replace numpy references with np
nums = [86,87,88,86,87,85,86]
sd = numpy.std(nums)
print(sd)

# high example
nums2 = [32,111,138,28,59,77,97]
sd = numpy.std(nums2)
print(sd)

# variance low
vr = numpy.var(nums)
print(vr)

# variance high
vr = numpy.var(nums2)
print(vr)
```

## Percentiles & Quartiles

Percentiles are a way to describe the position of a particular value within a dataset. In simple terms, a percentile tells you the percentage of values in a dataset that are less than or equal to a given value.

For example, suppose you have a dataset of exam scores for a class of students. If a student's score is at the 80th percentile, it means their score is higher than or equal to 80% of the scores in the dataset. In other words, they performed better than 80% of their peers.

Percentiles help provide a relative measure of where a specific value stands in relation to the entire dataset. They are often used to understand rankings, distributions, and to compare individual values against the larger context of the dataset.

The most commonly used percentiles are quartiles, which divide the dataset into four equal parts: the 25th percentile (Q1), the median (50th percentile or Q2), and the 75th percentile (Q3). These quartiles help understand the spread and distribution of the dataset by dividing it into four segments.


```python
# 75 percentile (75% are below)

import numpy

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = numpy.percentile(ages, 75)

print(x)
```

In the example below, we have the scores list representing an example dataset of exam scores.

We use the np.percentile() function from the numpy library to calculate the percentiles. The first argument is the dataset (scores), and the second argument is the desired percentile (25, 50, or 75 in this case).

We calculate the 25th percentile (Q1), median (Q2), and 75th percentile (Q3) and store them in variables p25, median, and p75, respectively.

Finally, we print the results using print(), displaying the calculated percentiles.


```python
import numpy as np

# Example dataset
scores = [60, 70, 75, 80, 85, 90, 95, 100]

# Calculate percentiles
p25 = np.percentile(scores, 25)  # 25th percentile (Q1)
median = np.percentile(scores, 50)  # 50th percentile (median or Q2)
p75 = np.percentile(scores, 75)  # 75th percentile (Q3)

# Print the results
print("25th percentile (Q1):", p25)
print("Median (Q2):", median)
print("75th percentile (Q3):", p75)
```

## Percentages

If you have a categorical data series and want to count the occurrences of each category and calculate the percentage of occurrence, you can use the value_counts() method from the pandas library.\
In this example, we have the categories list representing the categorical data series.

We create a pandas Series series from the categories list.

Using the value_counts() method on the Series, we calculate the count of occurrences for each unique category in the data series.

Next, we calculate the percentage by dividing the count of each category by the total length of the series and multiplying by 100.

Finally, we iterate over the index, count, and percentage of the value_counts series using a loop, and print the results, displaying the category, count, and percentage for each unique category.



```python
import pandas as pd

# Example categorical data series
categories = ['A', 'B', 'C', 'B', 'B', 'A', 'C', 'C', 'B']

# Create a pandas Series from the categories list
series = pd.Series(categories)

# Calculate the count and percentage for each category
value_counts = series.value_counts()
percentages = (value_counts / len(series)) * 100

#basic print
print (value_counts)
print(percentages)

# Print the results formatted
print("Value Counts and Percentages:")
for category, count, percentage in zip(value_counts.index, value_counts, percentages):
    print(f"{category}: Count={count}, Percentage={percentage:.2f}%")
```

## Z-scores outlier detection

One of the easiest ways to identify outliers in a data series is by using the z-score method. The z-score measures how many standard deviations a data point is away from the mean. If a data point has a z-score greater than a certain threshold (typically 2 or 3), it is considered an outlier.\
In this example, we have the data list representing a data series.

We calculate the mean (mean) and standard deviation (std) of the data series using np.mean() and np.std() from the numpy library.

Next, we set a threshold value (threshold) to determine when a data point is considered an outlier. Typically, a threshold of 2 or 3 SD is used, but you can adjust it according to your requirements.

We calculate the z-scores for each data point by subtracting the mean and dividing by the standard deviation.

We identify outliers by selecting data points whose absolute z-scores exceed the threshold using a list comprehension.

Finally, we print the identified outliers using print().


```python
import numpy as np

# Example data series
data = [10, 12, 15, 13, 11, 100, 14, 16, 9, 12, 17]

# Calculate the z-scores
mean = np.mean(data)
std = np.std(data)
threshold = 2  # 2 standard deviations from mean

z_scores = [(x - mean) / std for x in data]

# Identify outliers
outliers = [data[i] for i, z in enumerate(z_scores) if abs(z) > threshold]

# Print the outliers
print("Outliers:", outliers)
```

## Analysis (univariate vs bivariate)

In **univariate** analysis, a single variable is analyzed to understand its distribution, central tendency, spread, and other relevant statistics. It provides insights into the characteristics of individual variables.

**Bivariate** analysis, on the other hand, involves the analysis of two variables to explore the relationship between them. Common techniques used in bivariate analysis include correlation, scatter plots, and cross-tabulations (one variable is represented along the rows, and the other variable is represented along the columns, includes pivot tables).

When you extend the analysis to three or more variables simultaneously, it falls under the umbrella of **multivariate analysis**.

## correlation coefficients

Correlation coefficients are statistical measures that quantify the strength and direction of the linear relationship between two variables. The correlation coefficient ranges from -1 to 1, where:

1 indicates a perfect positive correlation (as one variable increases, the other increases proportionally).
0 indicates no linear correlation (the variables are not related).
-1 indicates a perfect negative correlation (as one variable increases, the other decreases proportionally). \

The output will be a correlation coefficient value between -1 and 1, indicating the strength and direction of the linear relationship between the variables "x" and "y."

```python
import numpy as np
from scipy.stats import pearsonr

# Sample data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 4, 5, 4, 5, 7])

# Calculate correlation coefficient and p-value (see below)
correlation_coefficient, p_value = pearsonr(x, y)

print("Correlation coefficient:",correlation_coefficient)
```

The correlation coefficient measures the strength and direction of the linear relationship between two variables. It is denoted by the symbol "r" and ranges between -1 and 1.

Positive Correlation (0 < r < 1): A positive correlation means that as one variable increases, the other variable tends to increase as well. A correlation coefficient close to 1 indicates a strong positive relationship.

Negative Correlation (-1 < r < 0): A negative correlation means that as one variable increases, the other variable tends to decrease. A correlation coefficient close to -1 indicates a strong negative relationship.

No Correlation (r = 0): A correlation coefficient of 0 indicates no linear relationship between the two variables. It doesn't mean that there is no relationship at all; it just means that the relationship is not linear.

The correlation coefficient is used to measure the linear relationship between two continuous variables. It is important to note that correlation does not imply causation. Even if two variables have a strong correlation, it does not necessarily mean that one causes the other.





## direction of correlation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Sample data for positive correlation
x1 = np.array([1, 2, 3, 4, 5, 6])
y1 = np.array([2, 4, 5, 4, 5, 7])

# Sample data for negative correlation
x2 = np.array([1, 2, 3, 4, 5, 6])
y2 = np.array([7, 6, 5, 4, 3, 2])

# Calculate correlation coefficients and p-values
correlation_coefficient1, p_value1 = pearsonr(x1, y1)
correlation_coefficient2, p_value2 = pearsonr(x2, y2)

# Scatter plot for positive correlation
plt.scatter(x1, y1, color='b', label=f"Positive Correlation: {correlation_coefficient1:.2f}")

# Scatter plot for negative correlation
plt.scatter(x2, y2, color='r', label=f"Negative Correlation: {correlation_coefficient2:.2f}")

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Positive and Negative Correlation')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

## Levels of Significance

When interpreting the correlation coefficient, it is essential to consider the level of significance, often denoted by "p-value." The p-value helps determine whether the observed correlation is statistically significant or simply due to random chance.

In hypothesis testing, the null hypothesis (H0) assumes that there is no correlation between the two variables (correlation coefficient = 0). The alternative hypothesis (H1) assumes that there is a correlation (correlation coefficient ≠ 0).

Common levels of significance are 0.05 (5%) and 0.01 (1%). If the p-value is less than the chosen level of significance (e.g., p < 0.05), then we reject the null hypothesis and conclude that there is a statistically significant correlation between the variables. If the p-value is greater than or equal to the level of significance, we fail to reject the null hypothesis, suggesting that there is not enough evidence to claim a significant correlation.

For example, if you calculate a correlation coefficient of 0.8 between two variables and the associated p-value is 0.03 (p < 0.05), you can conclude that there is a statistically significant positive correlation between the variables at the 5% level of significance.

```python
import numpy as np
from scipy.stats import pearsonr

# Sample data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 4, 5, 4, 5, 7])

# Calculate correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(x, y)

print("Correlation coefficient:",correlation_coefficient)
print("P-value:",p_value )

if p_value < 0.05:
    print("The correlation is statistically significant at the 5% level.")
else:
    print("There is no statistically significant correlation.")
```
