#Linear Regression

Scatter plots are used to visualize the relationship or pattern between two variables, typically referred to as the "independent variable" (or x-axis) and the "dependent variable" (or y-axis).

##Linear Regression Basic Postive

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

#type cast to int - the default is float
x = np.array([5,6,7,9,10,11,13,15,16,17,18,19,20])
y = np.array([85,87,89,90,95,98,100,102,103,105,107,109,110])

#linear regression metrics
#Perform linear regression using scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print("R-value:", r_value)
print("R-squared:", r_value**2)

#create scatter plot with line of regression
plt.scatter(x, y, label='Data')
plt.plot(x, slope * x + intercept, 'r', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Linear Regression')


```

## Linear Regression with Random Data

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set the random seed for reproducibility
np.random.seed(42)

# Generate two sets of 30 random numbers
x = np.random.rand(30)
y = 2 * x + 1 + np.random.randn(30) * 0.2  # Simulating a linear relationship with some noise

# Perform linear regression using scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Print the regression results
print("R-value:", r_value)
print("R-squared:", r_value**2)

# Create a scatter plot
plt.scatter(x, y, label='Data')
plt.plot(x, slope * x + intercept, 'r', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

## Values explained
**Slope (m):**
The slope represents the coefficient of the independent variable (X) in the linear regression equation. In this example, the slope is approximately 1.7214. It tells us how much the dependent variable (Y) is expected to change for a one-unit increase in the independent variable (X). A positive slope means that as X increases, Y is expected to increase as well. In this case, for every one unit increase in X, Y is expected to increase by approximately 1.7214 units.

**Intercept (b):**
The intercept represents the constant term in the linear regression equation. In this example, the intercept is approximately 1.0858. It is the value of the dependent variable (Y) when the independent variable (X) is zero. In some cases, this value may have no practical interpretation, especially if X cannot take values close to zero.

**R-value:**
R valuse is a measure of how well the regression line fits the data points. It ranges from -1 to 1 indicating negative or positive slant to the line of regression.

**R-squared:**
R-squared (R) is a measure of how well the regression line fits the data points. It ranges from 0 to 1, where 0 indicates that the regression line explains none of the variance in the data, and 1 indicates that the regression line perfectly fits the data points. In this example, R-squared is approximately 0.8948, which means that about 89.48% of the variance in Y can be explained by the linear relationship with X. A higher R-squared value indicates a better fit of the regression line to the data.

**p-value:**
The p-value measures the statistical significance of the slope coefficient. In this example, the p-value is very close to zero (3.2127861504964335e-15), which means there is strong evidence to reject the null hypothesis that the true slope is zero. In other words, the p-value suggests that the relationship between X and Y is statistically significant. A smaller p-value (typically less than 0.05) indicates a more significant relationship.

**Standard Error:**
The standard error represents the standard deviation of the residuals (the differences between the actual Y values and the predicted Y values from the regression line). In this example, the standard error is approximately 0.1116. A lower standard error indicates that the regression line is a better fit to the data.

In summary, the regression coefficients tell us about the strength and direction of the linear relationship between the variables. The slope indicates the rate of change of Y with respect to X, the intercept gives the value of Y when X is zero (if meaningful), R-squared shows how well the model fits the data, the p-value indicates the statistical significance of the relationship, and the standard error measures the accuracy of the predictions made by the regression line.



##Linear Regression Prediction

Remember - the "independent variable" (x-axis) and the "dependent variable" (y-axis).

```python
# predicting the future
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

#data series positive example
x = np.array([5,6,7,9,10,11,13,15,16,17,18,19,20])
y = np.array([85,87,89,90,95,98,100,102,103,105,107,109,110])

#data series negative example
#x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
#y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Perform linear regression using scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

#function to calculate predicted valued using slope and intercept
def myfunc(x):
  return slope * x + intercept

#function call
mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)

plt.show()

answer = myfunc(10)
print("The value predicted for y when x is 10 is ", answer)
# Print the regression results
print("R-value:", r_value)
print("R-squared:", r_value**2)
```

##Real World Example - NFL data set

An NFL (National Football League) dataset is a collection of structured information and statistics related to American football games, players, and teams within the NFL. Such datasets provide valuable insights for various purposes, including sports analytics, fantasy football analysis, and research. This dataset contains a range of columns including; \
team1 - home team \
team2 - away team\
score1 - home team score\
score2 - away team score\
QB1- home quarterback\
QB2 - away quarterback\
The term Elo - each NFL team is assigned an initial Elo rating at the start of a season. This initial rating is usually set based on historical performance, team strength, and various factors that the rating system's designers consider relevant.   the outcome of the game affects their respective Elo ratings.

```python
# predicting the future
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

# Load the dataset
df = pd.read_csv('nfl_elo_latest.csv')

#set NFL data set columns
x = df['qb1_value_pre']
y = df['qb1_value_post']

# Perform linear regression using scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Print the regression results
print("R-value:", r_value)
print("R-squared:", r_value**2)

# Create a scatter plot
plt.scatter(x, y, label='Data')
plt.plot(x, slope * x + intercept, 'r', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
